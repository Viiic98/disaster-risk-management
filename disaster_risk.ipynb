Check that values in a column are only numbers to finally convert the column to integer

```python
import pandas as pd
def check_rows(column):
  """ Remove special character from values in a column
      and let them with only numbers
      Return: column parsed to integer
  """
  for i, x in enumerate(df[column]):
    if ',' in x:
      df[column][i] = x.replace(',', '')
    if '.' in x:
      df[column][i] = x.replace('.', '')
    if not x.isdecimal():
      df[column][i] = '0'
  return pd.to_numeric(df[column], downcast='integer')
df['DEAD'] = check_rows('DEAD')
df['INJURED'] = check_rows('INJURED')
df['MISSING'] = check_rows('MISSING')
df['PERSONS'] = check_rows('PERSONS')
df['FAMILIES'] = check_rows('FAMILIES')
df['DESTROYED HOUSES'] = check_rows('DESTROYED HOUSES')
df['DAMAGED HOUSES'] = check_rows('DAMAGED HOUSES')
df['TRACKS'] = check_rows('TRACKS')
df['VEHICULAR BRIDGES'] = check_rows('VEHICULAR BRIDGES')
df['PEDESTRIAN BRIDGES'] = check_rows('PEDESTRIAN BRIDGES')
df['AQUEDUCT'] = check_rows('AQUEDUCT')
df['SEWERAGE'] = check_rows('SEWERAGE')
df['HEALTH CENTER'] = check_rows('HEALTH CENTER')
df['SCHOOL'] = check_rows('SCHOOL')
df['COMMUNITY CENTER'] = check_rows('COMMUNITY CENTER')
```

Normalize values removing special characters like "" using python

```python
def simplify(text):
	import unicodedata
	try:
		text = unicode(text, 'utf-8')
	except NameError:
		pass
	text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode("utf-8")
	return str(text)

for i, x in enumerate(df['CITY']):
  if x == None:
    df.drop(i, axis=0, inplace=True)
    continue
  df['CITY'][i] = simplify(x)
```

**Code to move the dataframe to S3 from SageMaker in order to download the dataframe as csv**

```python
from io import StringIO # python3; python2: BytesIO 
import boto3
def _write_dataframe_to_csv_on_s3(DESTINATION, dataframe, filename):
    """ Write a dataframe to a CSV on S3 """
    print(
        "Writing {} records to '{}' in bucket '{}'.".format(
            len(dataframe), filename, DESTINATION
        )
    )
    # Create buffer
    csv_buffer = StringIO()
    # Write dataframe to buffer
    dataframe.to_csv(csv_buffer, sep=",", index=False)
    # Create S3 object
    s3_resource = boto3.resource("s3")
    # Write buffer to S3 object
    s3_resource.Object(DESTINATION, filename).put(Body=csv_buffer.getvalue())
_write_dataframe_to_csv_on_s3('sagemaker-us-east-2-710303918543', df, 'data.csv')
```

`list_events`: list of lists, every list in list_events contains `event`, `number_of_occurences`. <br />
